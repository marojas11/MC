```python
import datetime
import time
solar=genfromtxt("monthrg.dat")
date=[]
manchas=[]
dias=solar[:,2]
years=[]
y=[]
for i in range(len(dias)):
    if dias[i]>0 and solar[i,0]>1900:
        date=datetime.date(int (solar[i][0]),int (solar[i][1]),int (solar[i][2]))
        manchas.append(solar[i][3])
        years.append(date)
        y.append(solar[i][3])
```

```python
from scipy.fftpack import fft, fftfreq
```

```python
fftm=abs(fft(manchas))

N = len(manchas) # Número de puntos de muestreo
f = 1./100 # Frecuencia
dt = 1./f / N # tiempo entre puntos de muestreo
t = np.linspace(0,1/f,N)
freq = fftfreq(N, dt)

plt.plot(freq,fftm)
xlim(-1,1)
```

```python
figure(figsize(20,10))
fftm=fft(y)
plot (years,y , c='gray')

fftm[np.abs(freq)>0.2] = 0
y_cleaned=ifft(fftm)

plot(years,np.real(y_cleaned),label=u'Señal filtrada', c='r')
legend()

ylim(0,200)
savefig("cicloSolar.png")
```
