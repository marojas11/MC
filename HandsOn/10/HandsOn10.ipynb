```python
solar=genfromtxt("monthrg.dat")
years=[]
manchas=[]
dias=solar[:,2]
y=[]
for i in range(len(dias)):
    if dias[i]>1 and solar[i,0]>1900:
        manchas.append(solar[i][3])
        years.append(solar[i][0])
        y.append(solar[i][3]/solar[i][1])
        
print len(years)
  
```

```python
from scipy.fftpack import fft, fftfreq
```

```python
fftm=abs(fft(manchas))

N = len(manchas) # Número de puntos de muestreo
f = 1./100.  # Frecuencia
dt = 1./f / N # tiempo entre puntos de muestreo
t = np.linspace(0,1/f,N)
freq = fftfreq(N, dt)
for i in range(len(fftm)):
    plt.plot([freq[i],freq[i]],[0,fftm[i]],"-")
plt.plot(freq,fftm,"o")
xlim(-1*10**4/2,1*10**4/2)
```
