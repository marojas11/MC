```python
def f(t):
    return exp(-t)*cos(2*pi*t)
t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)
figure(1)
#posición del subplot
subplot(211)
plot(t1, f(t1),'bo', t2, f(t2),'k')
subplot(212)
plot(t2, np.cos(2*np.pi*t2),'r--')
show()
```
