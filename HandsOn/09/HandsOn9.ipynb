```python
lena=misc.lena()
def inter(i1,j1,i2,j2):
    a=lena[i1,j1]
    b=lena[i2,j2]
    lena[i1,j1]=b
    lena[i2,j2]=a
    
for i in range(len(lena)):
    for j in range(len(lena)/2):
        inter(i,j,i,len(lena)-1-j)

figure(figsize=(10,10))
imshow(1.5*lena,cmap=plt.cm.gray)       
```
