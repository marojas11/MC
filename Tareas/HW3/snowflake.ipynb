En primer lugar se procede a construir las funciones que permitan generar los fractales, para ello se divide el problema en dos partes. Inicialmente se contruye la función fractales que permite generar sobre un solo lado de la figura inicial y los nuevos lados las nuevas figuras triangulares. Luego se utiliza la función copo para definir el triángulo inicial y llamar las iteraciones sobre cada uno de sus lados.

Para la realización de las siguientes funciones me basé en el tutorial de [Curva Y Copo De Nieve De Von Koch Con Python 3](http://linuxmusica.com/curva-y-copo-de-nieve-de-von-koch-con-python-3/)

```python
#Función que genera los fractales, los nuevos triágulos sobre lineas trazadas entre un punto inicial con coordenadas (xi,yi)
#y un punto final con coordenadas  (xf,yf). n corresponde al número de iteraciones 
def fractales(xi, yi, xf, yf, n):
    #Se tiene un número de iteraciones n, si no quedan iteraciones, se procede a montar la construcción sobre la gráfica
    if n == 0:
        plot([xi, xf], [yi, yf], lw=1.0, color='b')
    #Si quedan iteraciones disponibles se deben generar los puntos intermedios 
    elif n > 0:
        #Puntos intermedios o base del nuevo triángulo
        x1 = xi + (xf - xi) / 3.0
        y1 = yi + (yf - yi) / 3.0
        x3 = xf - (xf - xi) / 3.0
        y3 = yf - (yf - yi) / 3.0
        #El nuevo tamaño del triángulo con un radio a partir de la hipotenusa con los catetos introducidos
        radio = hypot(x3 - x1, y3 - y1)
        #Alfa corresponde al ángulo hallado a partir de los catetos
        alpha = atan2((y3 - y1), (x3 - x1))
        #Agrego pi/3 para generar el nuevo ángulo
        alpha += pi / 3.0
        #Hallo mi nuevo punto superior del triángulo pequeño
        x2 = x1 + radio * cos(alpha)
        y2 = y1 + radio * sin(alpha)
        
        #Repito el mísmo procedimiento para cada uno de los nuevos lados generados pero con una iteración menos
        fractales(xi, yi, x1, y1, n - 1)
        fractales(x1, y1, x2, y2, n - 1)
        fractales(x2, y2, x3, y3, n - 1)
        fractales(x3, y3, xf, yf, n - 1)


#Función que genera las coordenadas de los extremos del triángulo principal y llama la función que genera el fractal.
#Recibe como parámetro n el numero de iteraciones sobre el triángulo
def copo(n):
    #Se supone que el punto superior del triángulo corresponde al origen
    x_1 = 0 
    y_1 = 0
    #Para generar los puntos de los otros dos extremos se recurre a la función coseno y seno
    x_2 = -1 * cos(2 * pi / 3) #A 2/3 de vuelta en una circunferencia comenzando en el origen y con radio 1
    y_2 = -1 * sin(2 * pi / 3) 
    x_3 = -1 * cos(pi / 3)     #A 1/3 de vuelta en una circunferencia comenzando en el origen y con radio 1
    y_3 = -1 * sin(pi / 3)
    
    #Se generan los fractales en cada uno de los lados por medio de la función fractales
    fractales(x_1, y_1, x_2, y_2, n)
    fractales(x_2, y_2, x_3, y_3, n)
    fractales(x_3, y_3, x_1, y_1, n)

```

Ahora para generar la animación se procede a construir un ciclo que permita graficar el copo en cada una de sus etapas. Para ello utilizo la función copo y agrego las carácteristicas de la gráfica. Puede observar una animación al correr el recuadro, si no es así observar el gif


```python
#Ciclo para realizar las gráficas correspondientes a la animación hasta la 5ta iteración
for i in range(0,6):
    #Crear la figura nueva para cada iteración
    figure(figsize(10,10))
    #Generar dos módulos para reemplazar las imágenes 
    f, ax = plt.subplots()
    #Eliminar etiquetas y marcas de los ejes 
    tick_params(axis='x', which='both',  bottom='off',  top='off',  labelbottom='off')
    tick_params(axis='y', which='both',  right='off',  left='off',  labelleft='off')
    #limpiar
    plt.cla()
    #Esperar 1 segundo
    time.sleep(1)
    #Fijar los ejes 
    ax.set_xlim(-1, 1)
    ax.set_ylim(-3./2.*1, 1.0/2.0 )
    #Fijar la razon entre los ejes  
    ax.set_aspect(1.0)
    #Generar los números de iteración en el centro de la animación
    ax.text(-0.05,-0.6,r'$%d$'%i, fontsize=15)
    #Generar la figura
    copo(i)
    #Eliminar los bordes 
    ax.axis('off')
    #Limpiar la salida
    clear_output(wait=True)
    #Imprimir la siguiente linea
    display(f)
    #Guardar la imagen para generar el gif
    savefig("copo%d.png" %i)

plt.close()
```
