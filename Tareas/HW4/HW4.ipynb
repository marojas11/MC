**Difracción de Fraunhofer** El patrón de difracción producido producido por una apertura en el
régimen de Fraunhofer ( ver Born & Wolf, Principles of Optics capítulo 8 ) puede encontrarse
a través de una integral de Fourier y aproximarse numéricamente a través de una DFT. En
este ejercicio exploramos los patrones de difracción producidos por aperturas de diferentes
formas usando diferentes pupil functions (ver Born. pág 385). Imaginamos que las aperturas se
encuentran en el centro de una placa de lado L = 1 u y la representamos por medio de un array
cuadrado de 1024 elementos en cada dirección

a. Comenzando con un array de ceros de lado 1024 manipularlo con ciclos o usando slice notation para dejar todo en cero excepto elementos en un cuadrado de lado 32 en el centro.

```python
#Iniciar con el ambiente de numpy y matplotlib integrados
%pylab inline
```

```python
#Creación de la matriz de ceros de lado 1024
m=zeros([1024,1024])

```

```python
#Punto de partida para el reemplazo tanto en filas como en columnas
inicio=1024/2-32/2 
#Punto final para el reemplazo tanto en filas como en columnas
fin=1024/2+32/2
#Ciclo para realizar los reemplazos correspondientes en pro de generar el cuadrado de lado 32
for i in range(inicio,fin):
    for j in range(inicio,fin):
        #Reemplazo de las casillas correspondientes 
        m[i][j]=1

#Prueba de funcionamiento 
figure(figsize(5,5))
imshow(m)
```

b. Calcular la DFT del array resultante y aplicar sobre él la función `fftshift`, esto hace un cambio de coordenadas que arroja el patrón de interferencia esperado. De lo resultante tomar del centro un cuadrado de lado 200 y producir una imagen del logaritmo del módulo al cuadrado usando `imshow` con la representación de la apertura a su izquierda.

```python
#Importar funciones para el análisis de imágenes por medio de Fourier en dos dimensiones 
from scipy.fftpack import  fft2, fftshift
```

```python
#Asignación del tamaño de la figura
figure(figsize(10,10))
#Generación del panel izquierdo
subplot(1,2,1)
#Mostrar arreglo de partida
imshow(m)

#Aplicar transformada de Fourier en dos dimensiones al arreglo inicial
dftm=fft2(m)

#Aplicar cambio de coordenadas sobre la transformada obtenida
dfts=fftshift(dftm)

#Punto de partida para los datos mostrar en el panel derecho
inicio=1024/2-200/2
#Punto final para los datos mostrar en el panel derecho
fin=1024/2+200/2

#Lista que contendrá la matriz final de los datos de la transformada resaltada con logarítmo
b=[]  
#Ciclo que permita recorrer cada uno de los punto de la matriz a mostrar en el panel derecho de lado 200
for i in range(inicio,fin):
    #Lista que contendrá provisionalmente las filas del arreglo generadas en el siguiente ciclo
    a=[]
    for j in range(inicio,fin):
        #Generación de la lista aplicando el módulo (abs) a los valores de la transformada y resaltando con logaritmo, fue necesario aplicar +1 para que el logaritmo no generara errores con los valores en cero
        a.append(log((abs(dfts[i][j]))**2+1))
    b.append(a)
#Generación del panel derecho
subplot(1,2,2)
#Mostrar el arreglo de la transformada
imshow(b)

```

c. Repita los dos anteriores literales para una doble rendija cada una de ellas con
una altura de 200, grosor 20 y con sus ejes en las columnas 460 y 564. Esta vez tome un
cuadrado de lado 128 para la gráfica del patrón de difracción y no resalte con log

```python
#Generación de una nueva matriz de ceros
m=zeros([1024,1024])
#Punto de inicio para las columnas 
inicio=1024/2-200/2 
#Punto de fin para las columnas 
fin=1024/2+200/2

#Ciclo para generar las rendijas 
for j in range(inicio,fin):
    #Rendija de la izquierda
    for i in range(460-10,460+10):
       
        m[j][i]=1
    #Rendija de la derecha
    for i in range(564-10,564+10):
        
        m[j][i]=1

#Generación de la figura
figure(figsize(10,10))
#Panel izquierdo
subplot(1,2,1)
#Figura inicial de partida para las rendijas 
imshow(m)


#Aplicación de la transformada de Fourier en dos dimensiones  

dftm=fft2(m)
#Cambio de coordenadas 
dfts=fftshift(dftm)

#Parámetros para generar el panel derecho con las medidas sugeridas
inicio=1024/2-128/2
fin=1024/2+128/2

#Lista que contendrá los valores finales del módulo cuadrado de la transformada 
b=[]  
for i in range(inicio,fin):
    a=[]
    for j in range(inicio,fin):
        a.append((abs(dfts[i][j]))**2)
    b.append(a)

#Generación del panel derecho
subplot(1,2,2)
#Mostrar arreglo obtenido
imshow(b)
```

nuestro resultado con lo mostrado en la pág. 396 de los Principles de Born. I0 es la
amplitud máxima.


```python
#Inicialización de la placa a trabajar (matriz de ceros, 1024 de lado)
m=zeros([1024,1024])
#Generación de una matriz que contendrá la apertura circular obtenida
o=zeros([21,21])
#Generación de las posiciones en el arreglo que rodean un círculo de radio 10 casillas por medio de la función ogrid
#Recibe como parámetros los puntos de inicio y fin en las coordenadas en x (horizontal), luego en y (vertical)
x,y=ogrid[-10: 10+1, -10: 10+1]
#Por medio de una condición genero un arreglo de verdaderos y falsos que me permitirá reemplazar los puntos requeridos en la matriz de la apertura 
mask = (x)**2+(y)**2 <= 10**2
#Evalúo la función por medio de slice notation y reemplazo las casillas dentro del círculo por unos(1's) 
o[mask]=1

#Valores de inicio y fin para el reemplazo de la matriz de la apertura circular en el medio de la placa
inicio=1024/2-22/2 
fin=1024/2+20/2

#Ciclo para realizar el reemplazo correspondiente a la placa 
for j in range(inicio,fin):
    for i in range(inicio,fin):
        m[i][j]=o[i-inicio][j-inicio]

        
#Aplicación de la transformada sobre la matriz modificada en el ciclo anterior, la placa 
dftm=fft2(m)
#Cambio de coordenadas al centro
dfts=fftshift(dftm)

#Generación de abcisas y ordenadas para la gráfica correspondiente a las amplitudes obtenidas y descritas en el libro
x=2*pi*0.01*linspace(-512,512,1024)
y=(abs(dfts[1024/2]))/max((abs(dfts[1024/2])))

#Generación de las gráficas por medio de una figura inicial
plt.figure(figsize=(20,15))

#Generación del panel superior 
plt.subplot2grid((3,2),(0,0),colspan=2,rowspan=1)
#Gráfica de las abcisas y ordendas
scatter(x,y)
xlim(0,10)
ylim(0,1)
ylabel("$I/I_0$",size=16)
xlabel("$2\pi Rj/L$",size=16)

#Panel inferior izquierdo
plt.subplot2grid((3,2),(1,0),rowspan=2)
#Gráfica de la rendija inicial
imshow(m)

#Panel inferior derecho
plt.subplot2grid((3,2),(1,1),rowspan=2)
#Gráfica de la transformada obtenida resaltada con log
imshow(log((abs(dfts))**2+1))
plt.tight_layout()
plt.show()

```

```python
#Importar rutina de scipy para interpolar 
from scipy import interpolate
    
```

```python
#Generación de la función que interpola los puntos obtenidos de tipo cúbica, recibe como parámetros abcisas, ordenadas y el tipo
interpol_cubic = interpolate.interp1d(x,y,kind='cubic')
#Generación de la función que interpola los puntos obtenidos de tipo cúbica, recibe como parámetros abcisas, ordenadas y el tipo 
#pero ahora su reflexión en el eje x para hallar máximos
interpol_max = interpolate.interp1d(x,-y,kind='cubic')

```
