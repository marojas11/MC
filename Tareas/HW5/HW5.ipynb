## Punto a 
Resuelva las ecuaciones de movimiento para tres cuerpos de igual masa sujetos a su mutua atracción gravitacional con las siguientes condiciones iniciales. Tome G = 1 y use el método de Runge-Kutta de orden 4. Integre las ecuaciones de movimiento para media unidad de tiempo.


    
Ecuaciones:
    
$$r_1''=  G\frac{m_2}{||r_2-r_1||^3}(r_2-r_1)+ G\frac{m_3}{||r_3-r_1||^3}(r_3-r_1)=f(r_1,r_2,r_3)$$
$$r_2''=  G\frac{m_1}{||r_1-r_2||^3}(r_1-r_2)+ G\frac{m_3}{||r_3-r_2||^3}(r_3-r_2)=g(r_1,r_2,r_3)$$
$$r_3''=  G\frac{m_2}{||r_2-r_3||^3}(r_2-r_3)+ G\frac{m_1}{||r_1-r_3||^3}(r_1-r_3)=h(r_1,r_2,r_3)$$

Condiciones Iniciales:

$r_1(0) = ai + bj$

$r_1'(0) = ci + dj$

$r_2(0)= −ai − bj$

$r_2'(0) = ci + dj$

$r_3(0) = 0i + 0j$

$r_3'(0) = −2(ci + dj)$

Constantes 

$$a = 0.24250109$$
$$b = −0.0607718825$$
$$c = 0.93240737 $$
$$d = 0.86473146$$

```python
#Definir Constantes 
G=1 #Constante Gravitacional
m=1 #Masa de cada uno de los tres cuerpos 

#Condiciones Iniciales 
a = 0.24250109
b = -0.0607718825
c = 0.93240737 
d = 0.86473146

#Partícula 1
ri1 =array([a,b]) #Posición x,y
vi1= array([c,d]) #Vector velocidad x,y

#Partícula 2
ri2= array([-a,-b]) #Posición x,y
vi2 = array([c,d]) #Vector velocidad x,y

#Partícula 3
ri3= array([0.,0.]) #Posición x,y
vi3= -2*array([c,d]) #Vector velocidad x,y

#Visualización de las posiciones iniciales 
scatter(ri1[0],ri1[1]) #Particula 1
scatter(ri2[0],ri2[1]) #Partícula 2
scatter(ri3[0],ri3[1]) #Partícula 3


title("$Posiciones\  Iniciales$",size=16)
```

```python
# Aceleración sobre la primer particula
def f_(r1,r2,r3,v,t):
    global G
    #Vectores desplazamiento 
    r21=(r2-r1)
    r31=(r3-r1)
    return G*(m)*r21/((r21[0]**2+r21[1]**2)**(3./2))+G*(m)*r31/((r31[0]**2+r31[1]**2)**(3./2))

#  Aceleración sobre la segunda particula
def g(r1,r2,r3,v,t):
    global G
    r21=(r1-r2)
    r32=(r3-r2)
    return G*(m)*r21/((r21[0]**2+r21[1]**2)**(3./2))+G*(m)*r32/((r32[0]**2+r32[1]**2)**(3./2))

#  Aceleración sobre la tercera particula
def h(r1,r2,r3,v,t):
    global G
    r23=(r2-r3)
    r31=(r1-r3)
    return G*(m)*r31/((r31[0]**2+r31[1]**2)**(3./2))+G*(m)*r23/((r23[0]**2+r23[1]**2)**(3./2))

#Método de RungeKutta orden 4 para resolver las ecuaciones diferenciales para 3 cuerpos. Recibe como parámetro la fracción de tiempo a trabajar
def RungeKutta(step):
    #Lectura de variables de tiempo, posición y velocidad
    global t,r1,r2,r3,v1,v2,v3
    
    #Paso 1 de RungeKutta
    #Partícula 1
    k1=step*v1                   #Avance en Desplazamiento x=vt
    l1=step*f_(r1, r2, r3,v1,t)  #Avance en velocidad v=at
    #Partícula 2
    q1=step*v2                   #Avance en Desplazamiento                           
    m1=step*g(r1, r2, r3,v2,t)   #Avance en velocidad
    #Partícula 3
    n1=step*v3                   #Avance en Desplazamiento  
    p1=step*h(r1, r2, r3,v3,t)   #Avance en velocidad
    
    #Paso 2 de RungeKutta
    #Partícula 1
    k2=step*(v1+l1/2.)
    l2=step*f_(r1+k1/2., r2+q1/2., r3+n1/2., v1+l1/2.,t+step/2.)
    #Partícula 2
    q2=step*(v2+m1/2.)
    m2=step*g(r1+k1/2., r2+q1/2., r3+n1/2., v2+m1/2.,t+step/2.)
    #Partícula 3
    n2=step*(v3+p1/2.)
    p2=step*h(r1+k1/2., r2+q1/2., r3+n1/2., v3+p1/2.,t+step/2.)
    
    #Paso 3 de RungeKutta
    #Partícula 1
    k3=step*(v1+l2/2.)
    l3=step*f_(r1+k2/2., r2+q2/2., r3+n2/2., v1+l2/2.,t+step/2.)
    #Partícula 2
    q3=step*(v2+m2/2.)
    m3=step*g(r1+k2/2., r2+q2/2., r3+n2/2., v2+m2/2.,t+step/2.)
    #Partícula 3
    n3=step*(v3+p2/2.)
    p3=step*h(r1+k2/2., r2+q2/2., r3+n2/2., v3+p2/2.,t+step/2.)
    
    #Paso 4 de RungeKutta
    #Partícula 1
    k4=step*(v1+l3)
    l4=step*f_(r1+k3, r2+q3, r3+n3, v1+l3,t+step)
    #Partícula 2
    q4=step*(v2+m3)
    m4=step*g(r1+k3, r2+q3, r3+n3, v2+m3,t+step)
    #Partícula 3
    n4=step*(v3+p3)
    p4=step*h(r1+k3, r2+q3, r3+n3, v3+p3,t+step)
    
    #Actualización de velocidad y posición
    #Partícula 1
    r1+=(k1+2*k2+2*k3+k4)/6.
    v1+=(l1+2*l2+2*l3+l4)/6.
    #Partícula 2
    r2+=(q1+2*q2+2*q3+q4)/6.
    v2+=(m1+2*m2+2*m3+m4)/6.
    #Partícula 3
    r3+=(n1+2*n2+2*n3+n4)/6.
    v3+=(p1+2*p2+2*p3+p4)/6.
    #Tiempo
    t+=step
    
    
    
```

```python
#Método para implementar el valor del lapso de tiempo en base al número de Iteraciones numIter y el tiempo máximo tmax.
# Retorna un arreglo con el historial de posiciones 
def threeBody(numIter,tmax):
    
    #Lapso de Tiempo a trabajar
    step=tmax/numIter
    
    #Inicialización arreglo para almacenar posiciones
    history=np.array([[t,r1[0],r1[1],r2[0],r2[1],r3[0],r3[1]]])
    
    #Ciclo de Iteración para el método de RungeKutta
    for __ in range(numIter):
        #Aplicar rungekutta con el lapso de Tiempo calculado
        RungeKutta(step)
        #Almacenar las posiciones y velocidades en cada tiempo
        history=append(history,[[t,r1[0],r1[1],r2[0],r2[1],r3[0],r3[1]]],axis=0)
    
    return history
```

```python
#Método de RungeKutta orden 4 para resolver las ecuaciones diferenciales para 4 cuerpos. 
#Recibe como parámetro la fracción de tiempo a trabajar

def RungeKuttaC4(step):
    global t,r1,r2,r3,r4,v1,v2,v3,v4
    
    
    #Paso 1 de RungeKutta
    k1=step*v1                   #Avance en Desplazamiento x=vt
    l1=step*f_(r1,r2,r3,r4)      #Avance en velocidad v=at
    q1=step*v2
    m1=step*g(r1,r2,r3,r4)
    n1=step*v3
    p1=step*h(r1,r2,r3,r4)
    s1=step*v4
    w1=step*k(r1,r2,r3,r4)
    
    #Paso 2 de RungeKutta    
    k2=step*(v1+l1/2.)
    l2=step*f_(r1+k1/2., r2+q1/2., r3+n1/2.,r4+s1/2.)
    q2=step*(v2+m1/2.)
    m2=step*g(r1+k1/2., r2+q1/2., r3+n1/2.,r4+s1/2.)
    n2=step*(v3+p1/2.)
    p2=step*h(r1+k1/2., r2+q1/2., r3+n1/2.,r4+s1/2.)
    s2=step*(v4+w1/2.)
    w2=step*k(r1+k1/2., r2+q1/2., r3+n1/2.,r4+s1/2.)
    
    #Paso 3 de RungeKutta    
    k3=step*(v1+l2/2.)
    l3=step*f_(r1+k2/2., r2+q2/2., r3+n2/2., r4+s2/2.)
    q3=step*(v2+m2/2.)
    m3=step*g(r1+k2/2., r2+q2/2., r3+n2/2., r4+s2/2.)
    n3=step*(v3+p2/2.)
    p3=step*h(r1+k2/2., r2+q2/2., r3+n2/2., r4+s2/2.)
    s3=step*(v4+w2/2.)
    w3=step*k(r1+k2/2., r2+q2/2., r3+n2/2.,r4+s2/2.)
    
    #Paso 4 de RungeKutta
    k4=step*(v1+l3)
    l4=step*f_(r1+k3, r2+q3, r3+n3,r4+s3)
    q4=step*(v2+m3)
    m4=step*g(r1+k3, r2+q3, r3+n3, r4+s3)
    n4=step*(v3+p3)
    p4=step*h(r1+k3, r2+q3, r3+n3, r4+s3)
    s4=step*(v4+w3)
    w4=step*k(r1+k3, r2+q3, r3+n3, r4+s3)
    
    #Actualización de los vectores de velocidad y posición, y del tiempo
    r1+=(k1+2*k2+2*k3+k4)/6.
    v1+=(l1+2*l2+2*l3+l4)/6.
    r2+=(q1+2*q2+2*q3+q4)/6.
    v2+=(m1+2*m2+2*m3+m4)/6.
    r3+=(n1+2*n2+2*n3+n4)/6.
    v3+=(p1+2*p2+2*p3+p4)/6.
    r4+=(s1+2*s2+2*s3+s4)/6.
    v4+=(w1+2*w2+2*w3+w4)/6.
    
    t+=step

#Aceleración sobre la primer partícula
def f_(r1,r2,r3,r4):
    global G
    r21=(r2-r1)
    r31=(r3-r1)
    r41=(r4-r1)
    return G*(m)*r21/((r21[0]**2+r21[1]**2)**(3./2))+G*(m)*r31/((r31[0]**2+r31[1]**2)**(3./2))+G*(m)*r41/((r41[0]**2+r41[1]**2)**(3./2))

#Aceleración sobre la segunda partícula
def g(r1,r2,r3,r4):
    global G
    r12=(r1-r2)
    r32=(r3-r2)
    r42=(r4-r2)
    return G*(m)*r12/((r12[0]**2+r12[1]**2)**(3./2))+G*(m)*r32/((r32[0]**2+r32[1]**2)**(3./2))+G*(m)*r42/((r42[0]**2+r42[1]**2)**(3./2))

#Aceleración sobre la tercer partícula
def h(r1,r2,r3,r4):
    global G
    r23=(r2-r3)
    r13=(r1-r3)
    r43=(r4-r3)
    return G*(m)*r13/((r13[0]**2+r13[1]**2)**(3./2))+G*(m)*r23/((r23[0]**2+r23[1]**2)**(3./2))+G*(m)*r43/((r43[0]**2+r43[1]**2)**(3./2))

#Aceleración sobre la cuarta partícula
def k(r1,r2,r3,r4):
    global G
    r24=(r2-r4)
    r14=(r1-r4)
    r34=(r3-r4)
    return G*(m)*r14/((r14[0]**2+r14[1]**2)**(3./2))+G*(m)*r24/((r24[0]**2+r24[1]**2)**(3./2))+G*(m)*r34/((r34[0]**2+r34[1]**2)**(3./2))
```

```python
#Función que calcula el paso o lapso en el tiempo en base al número de iteraciones y el tiempo a trazar
def fourBody(numIter,tmax):
    
    step=tmax*1./numIter
    #Inicialización del arreglo para contener el tiempo, las componentes de las posiciones y de las velocidades 
    history=np.array([[t,r1[0],r1[1],r2[0],r2[1],r3[0],r3[1],r4[0],r4[1],v1[0],v1[1],v2[0],v2[1],v3[0],v3[1],v4[0],v4[1]]])
    #Ciclo para iterar el método de RungeKutta 
    for __ in range(numIter):
        RungeKuttaC4(step)
        #Almacenar cada una de las componentes en el arreglo
        history=append(history,[[t,r1[0],r1[1],r2[0],r2[1],r3[0],r3[1],r4[0],r4[1],v1[0],v1[1],v2[0],v2[1],v3[0],v3[1],v4[0],v4[1]]],axis=0)
    
    return history
```

$U = \sum_{1 \le i < j \le N} \frac{G m_i m_j}{\left\| \mathbf{q}_j - \mathbf{q}_i\right\|}$

$K = \sum_{i=1}^N \frac{ \left\|\mathbf{v}_i \right\| ^2 m_i}{2}$

```python
#Función que calcula la energía potencial para los cuatro cuerpos en base a los vectores obtenidos con a
def Pot(r10,r11,r21,r22,r31,r32,r41,r42):
    global G
    
    acum=0
    rarray=array([[0,0],[r10,r11],[r21,r22],[r31,r32],[r41,r42]])
    #Ciclo para evaluar la sumatoria de los 1<=i<j<=4 partículas
    for j in range(2,5):
        for i in range(1,j):
            acum+=G/((rarray[j][0]-rarray[i][0])**2+(rarray[j][1]-rarray[i][1])**2)**(0.5)
    
    return acum
#Función que calcula la energía cinética para los cuatro cuerpos en base a los vectores obtenidos con a
def Kin(v10,v11,v21,v22,v31,v32,v41,v42):
    acum=0
    varray=array([[0,0],[v10,v11],[v21,v22],[v31,v32],[v41,v42]])
    #Ciclo para evaluar la sumatoria para i<=4
    for i in range(1,5):
        acum+=((varray[i][0])**2+(varray[i][1])**2)/2.
    return acum
    
```

```python
#Leer el arreglo que contiene los valores de tiempo
tarray=history[:,0]
#Lista para los valores de la energía potencial
potential=[]
#Lista para los valores de la energía cinética 
kinetic=[]
#Lista para la suma total de las energías 
total=[]

#Ciclo para evaluar los valores obtenidos para las posiciones y las velocidades en las funciones de energía potencial y cinética
for t in range(len(tarray)):
    
    #Leer las posiciones 
    r10,r11,r21,r22,r31,r32,r41,r42=history[t,1:9]
    potential.append(-Pot(r10,r11,r21,r22,r31,r32,r41,r42))
                     
    #Leer las velocidades 
    v10,v11,v21,v22,v31,v32,v41,v42 = history[t,9:17]
    kinetic.append(Kin(v10,v11,v21,v22,v31,v32,v41,v42))
    #Sumar las anteriores 
    total.append(potential[t]+kinetic[t])
    
fig=figure(figsize(10,5))

plot(tarray,potential, label="Energia Potencial")
plot(tarray,kinetic, label="Energia Cinetica")
plot(tarray,total,label="Energia Total")
legend()
ylim(-20,15)


```
