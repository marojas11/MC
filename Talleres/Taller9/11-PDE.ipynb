## La Ecuación de difusión

$$D \frac{\partial^2f}{\partial x^2}=\frac{\partial f}{\partial t}$$

La derivada con respecto a $t$ se aproxima con una diferencia futura de acuerdo a:

$$\frac{\partial f}{\partial t} \approx \frac{f(t+\Delta t,x)-f(t,x)}{\Delta t}.$$

Y la segunda derivada con respecto a x:

$$
\frac{\partial^2 f}{\partial x^2} \approx \frac{f(t,x+\Delta x) - 2 f(t,x) + f(t,x-\Delta x)}{(\Delta x)^2}
$$

Se discretiza ahora el espacio en N puntos con espaciamiento $\Delta{x}$ y el tiempo en M puntos con espaciamiento $\Delta t$. De tal forma que

$$
t_{n}=t_\textrm{min}+\Delta t\ \ n\geq 0
$$

$$
x_{m} = x_\textrm{min}+m\Delta x\ \ 0\leq m\leq N - 1.
$$

Y resolviendo para $f(x_m,t_{n+1})$:

$$f(t_{n+1},x_m) = f(t_n,x_m) + D \frac{\Delta t}{\Delta x^2}\left( f(t_n,x_{m+1}) + f(t_n,x_{m-1}) - 2 f(t_n,x_m) \right)$$

con $\alpha:=D\frac{\Delta t}{\Delta x^2}$ < 1/2 para que el método sea computacionalmente estable.

### Condiciones de frontera periódicas

Tomar $f(0,x) = \frac{1}{\sqrt{2\pi}}e^{-x^2/2}$, usar condiciones de frontera periódicas, y fijar $D=1$.

```python
N=50 # posición
M=1000. # tiempo
tmin=0.
tmax=2.
xmin=-5.
xmax=5.
Delta_t=(tmax-tmin)/M
Delta_x=(xmax-xmin)/N
alpha=Delta_t/Delta_x**2
Y=np.zeros((M,N)) # Un renglón para cada tiempo
# Establecer la condición inicial
Y[0]=1./np.sqrt(2*np.pi)*np.exp(-np.linspace(xmin,xmax,N)**2/2)
print alpha
if alpha>0.5:
    print u"Atención alpha > 2"
for i in range(1,int(M)):
    Y[i] = Y[i-1] + alpha * (np.roll(Y[i-1],1)+np.roll(Y[i-1],-1)-2*Y[i-1])
```

```python
timeaxis=np.linspace(tmin,tmax,M)
spaceaxis=np.linspace(xmin,xmax,N)
plt.figure(figsize=(12,10))
plt.pcolor(spaceaxis,timeaxis,Y,cmap='gist_heat')
plt.xlabel("x")
plt.ylabel("t")
plt.xlim(xmin,xmax)
plt.colorbar()
plt.show()
```

```python
# Explorar la forma en la que aumenta la desviación estándard
desviaciones=np.apply_along_axis(lambda x: sum(spaceaxis**2*x),1,Y)
plt.figure(figsize=(10,5))
plt.plot(timeaxis,desviaciones)
plt.xlabel("t")
plt.ylabel("$\sigma^2$")
plt.show()
```

### Extremos absorventes

Resolvamos ahora la ecuación con $f(0,x) = 1/\sqrt{2\pi}e^{-x^2/2}$, usando condiciones de Dirichlet con $f(t,x_{\textrm{min}})=f(t,x_\textrm{max})=0$.

```python
N=50 # posición
M=1000. # tiempo
tmin=0.
tmax=2.
xmin=-5.
xmax=5.
Delta_t=(tmax-tmin)/(M-1)
Delta_x=(xmax-xmin)/(N-1)
alpha=Delta_t/Delta_x**2
Y=np.zeros((M,N)) # Un renglón para cada tiempo
# Establecer la condición inicial
Y[0]=1./np.sqrt(2*np.pi)*np.exp(-np.linspace(xmin,xmax,N)**2/2)
# Fijar las condiciones de frontera
Y[0,0]=0
Y[0,-1]=0
print alpha
if alpha>0.5:
    print u"Atención alpha > 2"
for i in range(1,int(M)):
    Y[i] = Y[i-1] + alpha * (np.roll(Y[i-1],1)+np.roll(Y[i-1],-1)-2*Y[i-1])
    Y[i,0]=0
    Y[i,-1]=0
```

```python
timeaxis=np.linspace(tmin,tmax,M)
spaceaxis=np.linspace(xmin,xmax,N)
plt.figure(figsize=(12,10))
plt.pcolor(spaceaxis,timeaxis,Y,cmap='gist_heat')
plt.colorbar()
plt.xlabel("x")
plt.ylabel("t")
plt.xlim(xmin,xmax)
plt.show()
```

```python
cantidad_total=np.apply_along_axis(sum,1,Y)
cantidad_total=cantidad_total/max(cantidad_total) # normalizar con respecto al máximo
plt.figure(figsize=(10,5))
plt.plot(timeaxis,cantidad_total)
plt.xlabel("t")
plt.ylabel("S")
plt.show()
```

## La Ecuación de Onda 1D

$\frac{\partial^2 \psi}{\partial x^2} = \frac{1}{v^2} \frac{\partial^2 \psi}{\partial t^2}$

### Condiciones de frontera periódicas

```python
tmin=0
tmax=.003105882
xmin=-0.33  # en metros
xmax=0.33   # en metros
Nt=1000
Nl=100
dt=(tmax-tmin)/Nt
dx=(xmax-xmin)/Nl
v=425. # m/s
r=v*dt/dx
print "r=",r
```

```python
#condiciones de frontera periódicas
xcoords=np.linspace(xmin,xmax,Nl)
tcoords=np.linspace(tmin,tmax,Nt)
chord=np.zeros((Nt,Nl))
# fijar la condición inicial
chord[0]=np.exp(-xcoords**2/(2*0.1**2))
chord[1]=chord[0]+r**2/2.*(np.roll(chord[0],1)+np.roll(chord[0],-1)-2*chord[0])
plt.figure(figsize=(10,3))
plt.plot(xcoords,chord[0])
plt.plot(xcoords,chord[1])
plt.show()
for i in range(2,Nt):
    ppchord=chord[i-2]
    pchord=chord[i-1]
    chord[i]=2*(1-r**2)*pchord-ppchord+r**2*(np.roll(pchord,1)+np.roll(pchord,-1))
plt.figure(figsize=(10,10))
plt.pcolor(xcoords,tcoords,chord,cmap='gist_heat')
plt.xlim(xmin,xmax)
plt.ylim(tmin,tmax)
plt.show()
```

### Extremos fijos

```python
tmin=0
tmax=.003105882
xmin=-0.33  # en metros
xmax=0.33   # en metros
Nt=1000
Nl=100
dt=(tmax-tmin)/Nt
dx=(xmax-xmin)/Nl
v=425. # m/s
r=v*dt/dx
print "r=",r
```

```python
#fija en los extremos
xcoords=np.linspace(xmin,xmax,Nl)
tcoords=np.linspace(tmin,tmax,Nt)
chord=np.zeros((Nt,Nl))
# fijar la condición inicial
std=0.05
chord[0]=0.01*np.exp(-xcoords**2/(2*std**2))
chord[0,0]=0
chord[0,-1]=0
chord[1]=chord[0]+r**2/2.*(np.roll(chord[0],1)+np.roll(chord[0],-1)-2*chord[0])
plt.figure(figsize=(10,3))
plt.plot(xcoords,chord[0])
plt.show()
for i in range(2,Nt):
    ppchord=chord[i-2]
    pchord=chord[i-1]
    chord[i]=2*(1-r**2)*pchord-ppchord+r**2*(np.roll(pchord,1)+np.roll(pchord,-1))
    chord[i,0]=0
    chord[i,-1]=0
plt.figure(figsize=(10,10))
plt.pcolor(xcoords,tcoords,chord,cmap='gist_heat')
plt.xlim(xmin,xmax)
plt.ylim(tmin,tmax)
plt.show()
```

```python
%matplotlib osx
from matplotlib import animation

fig=plt.figure(figsize=(12,1))
ax = fig.add_subplot(111, autoscale_on=False)
chordplot,=ax.plot([],[],"o",ms=2)
plt.xlim(xmin,xmax)
plt.ylim(-0.01,0.01)

time_template = 't = %.2f'
time_text = ax.text(0.5, 0.8, 'nada',transform=ax.transAxes)

def animate(i):
    chordplot.set_data(xcoords,chord[3*i])
    time_text.set_text(time_template%(i*dt*1000)+"ms")
    return chordplot,time_text

animacion = animation.FuncAnimation(fig, animate, np.arange(1,len(chord)/3),interval=1, blit=False)
plt.show()
```

## La Ecuación de Onda 2D

$\frac{\partial^2 \psi}{\partial x^2} + \frac{\partial^2 \psi}{\partial y^2} = \frac{1}{v^2} \frac{\partial^2 \psi}{\partial t^2}$

### Condiciones de frontera periódicas

```python
tmin=0
tmax=.3105882/2.
xmin=-0.33  # en metros
xmax=0.33   # en metros
ymin=-0.33
ymax=0.33
Nt=6000
Nl=30
dt=(tmax-tmin)/Nt
dx=(xmax-xmin)/Nl
dy=(ymax-ymin)/Nl
v=425. # m/s
r=v*dt/dx
print "r=",r
```

```python
#condiciones de frontera periódicas
xcoords=np.linspace(xmin,xmax,Nl)
ycoords=np.linspace(ymin,ymax,Nl)
xmesh,ymesh=np.meshgrid(xcoords,ycoords)
tcoords=np.linspace(tmin,tmax,Nt)
drum=np.zeros((Nt,Nl,Nl))
# fijar la condición inicial
stdev=0.1
drum[0]=np.exp(-xmesh**2/(2*stdev**2))*np.exp(-ymesh**2/(2*stdev**2))
drum[1]=drum[0]+r**2/2.*(np.roll(drum[0],1,axis=0)+np.roll(drum[0],-1,axis=0)+np.roll(drum[0],1,axis=1)+np.roll(drum[0],-1,axis=1)-4*drum[0])
# Mostrar la condición inicial
plt.figure(figsize=(10,3))
plt.imshow(drum[0],cmap='gist_heat')
plt.show()
```

```python
for i in range(2,Nt):
    pdrum=drum[i-1]
    ppdrum=drum[i-2]
    drum[i]=2*(1.-2.*r**2)*pdrum-ppdrum+r**2*(np.roll(pdrum,1,axis=1)+np.roll(pdrum,-1,axis=1)+np.roll(pdrum,1,axis=0)+np.roll(pdrum,-1,axis=0))
```

```python
# Animación
%matplotlib osx
from matplotlib import animation
# Placa fría con borde caliente

fig=plt.figure(figsize=(10,10))
plt.hold(True)
plt.pcolor(xcoords,ycoords,drum[0],cmap='gist_heat',vmin=0.,vmax=1.)
plt.xlim(xmin+dx,xmax-dx)
plt.ylim(ymin+dy,ymax-dy)
#plt.colorbar()

def animate(i):
    plt.pcolor(xcoords,ycoords,drum[i],cmap='gist_heat',vmin=0.,vmax=1.)
    plt.xlim(xmin+dx,xmax-dx)
    plt.ylim(ymin+dy,ymax-dy)
animacion = animation.FuncAnimation(fig, animate, np.arange(1,len(drum)),interval=20, blit=False)
plt.show()
plt.hold(False)
```

## Ecuación de Poisson 2D

$\frac{\partial^2 \Phi}{\partial x^2} + \frac{\partial^2 \Phi}{\partial y^2} = -\rho/\epsilon_0$

$\Phi_{i,j}=\frac{1}{4}(\Phi_{i+1,j}+\Phi_{i-1,j}+\Phi_{i,j+1}+\Phi_{i,j-1}+\frac{\Delta{x}^2}{\epsilon_0}\rho_{i,j})$

```python
%matplotlib inline
e0=8.85e-12 # unidades SI
q=1e-2 # en C
L=0.16 # en m
Nl=51 # 1 caja en la mitad 25 hacia la derecha y hacia la izquierda
xmin=-L/2
xmax= L/2
ymin=-L/2
ymax= L/2
xcoords=np.linspace(xmin,xmax,Nl)
ycoords=np.linspace(-ymin,-ymax,Nl)  # up is up
xmesh,ymesh=np.meshgrid(xcoords,ycoords)
Dx=(xmax-xmin)/Nl
density=np.zeros((Nl,Nl))
density[2,3:-3]=q/(L*(Nl-6)/Nl*L/Nl)             # carga positiva
density[-3,3:-3]=-q/(L*(Nl-6)/Nl*L/Nl)            # carga negativa
field=np.zeros((Nl,Nl))
NumIter=100000
```

```python
scaled_density=Dx**2*density
for __ in range(NumIter):
    field=1/4.*(np.roll(field,1,axis=0)+np.roll(field,-1,axis=0)+np.roll(field,1,axis=1)+np.roll(field,-1,axis=1)+scaled_density)
    field[0]=0.
    field[-1]=0.
    field[:,0]=0.
    field[:,-1]=0.
```

```python
plt.figure(figsize=(12,10))
plt.pcolor(xmesh,ymesh,field,cmap='gist_heat')
plt.colorbar()
plt.contour(xmesh,ymesh,field)
plt.xlim(xmin,xmax)
plt.ylim(ymin,ymax)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(ycoords,1e3*field[:,Nl/2],"o--")
plt.xlabel("$y/m$",fontsize=15)
plt.xlim(xmin,xmax)
plt.ylabel(r"$\Phi/mV$",fontsize=15)
plt.show()
```

```python
# Calcular el campo eléctrico
efieldx=np.diff(field,axis=1)/Dx
efieldy=np.diff(field,axis=0)/Dx
plt.figure(figsize=(10.,10.))
plt.streamplot(xcoords[1:],ycoords[1:],efieldx[1:,:],efieldy[:,1:])
plt.xlim(xmin,xmax)
plt.ylim(xmin,xmax)
plt.xlabel("x/m")
plt.ylabel("y/m")
plt.show()
```

## Ecuación de Schrödinger

$-\frac{\hbar^2}{2m}\frac{\partial^2 \psi}{\partial x^2} + V\psi = i\hbar \frac{\partial \psi}{\partial t}$

```python
%matplotlib osx
```

```python
L=1.
xmin=-L/2
xmax=L/2
d=0.4
tmin=0.
tmax=0.4
Nt=1000
Nl=50
dt=(tmax-tmin)/Nt
dx=(xmax-xmin)/Nl
A=1./np.sqrt(d)
xcoords=np.linspace(xmin,xmax,Nl)
erwin=np.zeros((Nt,Nl),dtype=np.complex)
erwin[0]=np.cos(xcoords*np.pi/L)+np.cos(3*xcoords*np.pi/L)+np.cos(5.*xcoords*np.pi/L)
#erwin[0][xcoords<=0]=0
#erwin[0][-1]=0
totprob=sum((np.abs(erwin[0])**2))*dx
erwin[0]=erwin[0]/np.sqrt(totprob)
erwin=np.matrix(erwin)
```

```python
plt.figure()
plt.plot(np.abs(np.array(erwin[0])**2)[0])
plt.show()
```

```python
idmat=np.diag(1.+np.zeros(Nl))
ham=-0.5/dx**2*(np.diag(1+np.zeros(Nl-1),1)+np.diag(-2+np.zeros(Nl),0)+np.diag(1+np.zeros(Nl-1),-1))
op=np.matrix(idmat+1.j*dt/2.*ham)
invop=np.linalg.inv(op)
```

```python
plt.figure()
plt.plot(np.sort(np.linalg.eigvals(ham)),"o")
plt.show()
```

```python
for i in range(1,Nt):
    erwin[i]=2*np.transpose(invop*np.transpose(erwin[i-1]))-erwin[i-1]
    erwin[i,0]=0.
    erwin[i,-1]=0.
    # normalizar
    totprob=sum((np.abs(np.array(erwin[i]))**2)[0])*dx
    #print totprob,
    erwin[i]=erwin[i]/np.sqrt(totprob)
```

```python
%matplotlib osx
from matplotlib import animation
fig=plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, autoscale_on=False, xlim=(xmin, xmax), ylim=(0, 5))
ax.grid(True) # Add grid

probability, = ax.plot([], [], '-', lw=2)

#legend()
#time_template = 'time = %.2f'
#time_text = ax.text(0.05, 0.9, '', transform=ax.transAxes)
#dt = 0.05

def animate(i):
    arerwin=np.array(erwin[i*10])
    probability.set_data(xcoords, np.abs(arerwin)**2)
    #time_text.set_text(time_template%(i*dt))
    return probability #time_text



animacion = animation.FuncAnimation(fig, animate, np.arange(1, len(erwin)/10),
    interval=100, blit=False)

plt.show()
```

## Schrodinger 1D - Oscilador Armónico (Energías)

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
Nl=101
xmin=-np.sqrt(Nl) # De lo contrario la cuadrícula es muy pequeña y pronto se vuelve un pozo infinito
xmax=np.sqrt(Nl)
dx=(xmax-xmin)/(Nl-1.)
xvals=np.linspace(xmin,xmax,Nl)
ham=np.matrix((0.5)*np.diag(xvals**2)+1./dx**2*np.identity(Nl)-0.5/dx**2*(np.diag(1.+np.zeros(Nl-1),-1)\
                                         +np.diag(1.+np.zeros(Nl-1),1)))
```

```python
energias=np.sort(np.linalg.eigvals(ham))
plt.figure(figsize=(10,6))
plt.xlabel(r"$n$",fontsize=20)
plt.ylabel(r"$E/\hbar\omega$",fontsize=20)
plt.plot(energias,"o")
plt.show()
```

```python
# En teoría la pendiente es uno y el corte con el eje 0.5
np.polyfit(range(20),energias[:20],1)
```

```python
eigsys=np.linalg.eigh(ham)
```
