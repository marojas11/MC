```python
a=0.1
b=a
c=14.
times=np.linspace(0,600,100000)
from scipy.integrate import odeint
```

```python
def Rossler(Y,t):
    global a,b,c
    Syst=array([-Y[1]-Y[2],Y[0]+a*Y[1],b+Y[2]*(Y[0]-c)])
    return Syst

funsol=odeint(Rossler,[1.,1.,1.],times,printmessg=True)
```
