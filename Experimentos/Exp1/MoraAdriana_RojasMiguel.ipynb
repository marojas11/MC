```python
#Importar paquetes
import csv
import scipy.stats
import numpy as np
```

```python
#1. 50 pt Giro de Italia: En el archivo giro_2014.csv hay datos de los tiempos de todos los corredeores en cada una
#de las 21 etapas del giro de Italia del 2014.
# PUNTO A


a=[] #Lista de filas 
nombre=[] #Lista de Nombres
e2=[] #Lista tiempos de la etapa 2 acumulada
e21=[]#Lista tiempos de la etapa 21 acumulada
e4solo=[]#Lista tiempos solo de la etapa 4
e21solo=[]#Lista tiempos solo de la etapa 21
c=0 
acum=0

#Importar Archivo
with open("giro_2014.csv", 'rb') as csvfile: #importar desde csv.reader
    spamreader = csv.reader(csvfile, delimiter=",", quotechar='|') 
    for row in spamreader: 
        a.append(row) #agregar a las filas
    for i in range (1,len(a)): 
        b=a[i] 
        for j in range (len(b)):
            
            #Nombres
            if(j==0): 
                nombre.append(b[j]) #agregar a las filas
            #Etapa 2
            if(j==4): 
                c=int(b[j])
                
                e2.append(c) #agregar a las filas
            #Etapa 21
            if(j==23): 
                if b[j]=="NA":
                    b[j]=0
                else:
                    e21.append(int(b[j])) #agregar a las filas
                    e21solo.append(int(b[j])-int(b[j-1])) #agregar a las filas
            #Etapa 4
            if(j==6):
                if b[j]=="NA":
                    b[j]=0
                else:
                    e4solo.append(int(b[j])-int(b[j-1])) #agregar a las filas
            
   
```

```python
#2.Doble distribución normal

#PUNTO A
#Crear numeros con distribución normal alrededor de una media de 10 y con desviación estandar 2
x = np.random.normal(10, 2, 10000)
#Crear numeros con distribución normal alrededor de una media de 20 y con desviación estandar 2
y = np.random.normal(20, 2, 10000)
#Unir distribuciones
d=append(x,y)
#Histograma
hist(d, normed=True, bins=25)
figure(figsize(10,5))
```

```python
#PUNTO B
#Histograma acumulado
q=hist(d, normed=True, bins=25, cumulative=True)
```

```python
#PUNTO C


qq=q[0]#Valores en el eje y
qx=q[1]#Valores en el eje x


#Función que recibe por parametro el número de números a generar
def random_2normal(l):
    j=[] #Lista para los numeros aleatorios
    retorno=[] #Lista para los valores del eje x 
    comparar=1000 #Valor inicial para la comparacion
   
   
    #Ciclo para asignar los valores aleatorios
    for x in range (l):
        numeros=random.random()
       
        j.append(numeros)
    #Ciclo para comparar los valores generados y el eje y
    for s in range(len(j)):
        #Valor de la variable aleatoria
        var=j[s]
        #Ciclo para comparar los valores generados y el eje y
        for n in range(len(qq)):
            #Diferencia de los valores 
            funcion= abs(var-qq[n])
            #Minimizar error
            if(funcion<comparar):
                comparar=funcion
                pp=n
        #aqui guardarara el arreglo para los numeros en el eje x
        retorno.append(qx[pp])   
    return retorno
        
                
                

#Distribución para 1000 numeros

x=random_2normal(1000)

        
#Histograma
hist(x,normed=True,bins=100,cumulative=True)

print "Example"
```
